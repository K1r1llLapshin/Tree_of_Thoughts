```python
from IPython.display import display, Markdown
from core.client import LLMClient
from core.search_algorithm.bfs import BFS
```

```python
prompt = '''
Напиши план по уборке дома, который разбит на задачи и подзадачи. Каждая задача должна быть описана подробно.План должен быть логичным и последовательным, 
чтобы уборка проходила максимально эффективно.
'''
# Инициализация клиента
client = LLMClient(model_name="deepseek-chat", temperature=0.7)

# Инициализация алгоритма поиска
search_algorithm = BFS(client, count_thoughts=3, breadth_limit=2, max_depth=3)

# Запуск поиска
result = search_algorithm.solve(prompt)
```

```python
display(Markdown(result))
```
