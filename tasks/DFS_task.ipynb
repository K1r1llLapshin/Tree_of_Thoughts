```python
from core.client import LLMClient
from core.search_algorithm.dfs import DFS
from IPython.display import display, Markdown
```

```python
prompt = '''
Дано уравнение, где каждая буква — это уникальная цифра от 0 до 9:
$$SEND + MORE = MONEY$$
Цель: Найти такие значения букв, чтобы равенство стало верным.
'''
# Инициализация клиента
client = LLMClient(model_name="deepseek-chat", temperature=0.7)

# Инициализация алгоритма поиска
search_algorithm = DFS(client, count_thoughts=4, max_depth=7, threshold=0.6)

# Запуск поиска
result = search_algorithm.solve(prompt)
```

```python
display(Markdown(result))
```
